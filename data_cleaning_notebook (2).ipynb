```python
import pandas as pd



df = pd.read_csv('synthetic_retail_sales.csv')

# 1. remove duplicate rows
df = df.drop_duplicates()

# 2. clean Region column - remove spaces and fix casing
df['Region'] = df['Region'].str.strip().str.lower()
df['Region'] = df['Region'].replace({'nort': 'north', 'centrl': 'central'})
df['Region'] = df['Region'].str.title()

# 3. convert Date to proper date type
df['Date'] = pd.to_datetime(df['Date'])

# 4. fix negative prices - make them positive
df['Price'] = df['Price'].abs()

# 5. fill missing Quantity and Price with column average
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())

df.to_csv('cleaned_output.csv', index=False)
print(df.shape)
```
